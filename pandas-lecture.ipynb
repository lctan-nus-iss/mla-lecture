# Panda Series

```python
# creates a Pandas Series - a 1D array
# with associated indexes
s = pd.Series([21, 32, 45])

# s =
# 0  21
# 1  32
# 2  45
print('s =\n', s,sep='')
print(s[0])

# update series by index-name
s[1] = 38
print('s =\n', s,sep='')
```

### A Pandas Series can be manipulated with index names 


```python
# adding indexes to our series
s = pd.Series([21, 32, 45], 
        index=['mary', 'john', 'james'])

# s =
# mary   21
# john   32
# james  45
print('s =\n', s, sep='')

# ['mary', 'john', 'james']
print('index =', s.index)

# update series by index-name
s['john'] = 38
print('s =\n', s, sep='')

#s.index[2] ='jason' #index are immutable 
#print('s =\n', s, sep='')
```

### Panda Series accessed by position or name 

```python
# s.iloc[1] = 32
print('s.iloc[1] =', s.iloc[1])

# s.iloc[1:1] = []
print('s.iloc[1:1] =', s.iloc[1:1])

# s.iloc[1:2] = john 32
print('s.iloc[1:2] =', s.iloc[1:2])

# s.iloc[1:3] =
# john  32
# james 45
print('s.iloc[1:3] =\n', s.iloc[1:3], sep='')

# s.loc['john'] = 32
print("s.loc['john'] =", s.loc['john'])

# s.loc['john':'john'] = john 32
print("s.loc['john':'john'] =", 
    s.loc['john':'john'])

# s.loc['john':'james'] =
# john     32
# james    45
print("s.loc['john':'james'] =\n", 
    s.loc['john':'james'], sep='')
```

# Dataframe

```python
# creates a Pandas DataFrame - a 2D array
# with default index and column names 
df = pd.DataFrame([21, 32, 45])

# df =
#     0
# 0  21
# 1  32
# 2  45
print('df =\n', df, sep='')
```

## Specifying Index Names and Column Names to a DataFrame

```python
# creates a dataframe with column 
# and index names
df = pd.DataFrame([21, 32, 45],
    columns = ['age'],
    index = ['mary', 'john', 'james']    
)

# df =
#       age
# mary  21
# john  32
# james 45
print('df =\n', df, sep='')
print(df.head())
print(df.to_string())
print(df)
```

### A more complex example

```python
# creates a more complex dataframe
df = pd.DataFrame(
    [
        [21, 1.7, 55],
        [32, 1.8, 65],
        [45, 1.6, 50]
    ],
    columns = ['age', 'height', 'weight'],
    index = ['mary', 'john', 'james']    
)

# df =       
#        age  height  weight
# mary    21     1.7      55
# john    32     1.8      65
# james   45     1.6      50
print('df =\n', df, sep='')
```

### Accessing elements in Pandas by position

```python
# creates a pandas dataframe
df = pd.DataFrame(
    [
        [21, 1.7, 55],
        [32, 1.8, 65],
        [45, 1.6, 50]
    ],
    columns = ['age', 'height', 'weight'],
    index = ['mary', 'john', 'james']    
)

# df =       
#        age  height  weight
# mary    21     1.7      55
# john    32     1.8      65
# james   45     1.6      50
print('df =\n', df, sep='')

# retrieving as a pandas series
# df.iloc[1] =
# age       32.0
# height     1.8
# weight    65.0
print('df.iloc[1] =\n', df.iloc[1], sep='')

# retrieving as a pandas dataframe
# df.iloc[[1]] =
#       age  height  weight
# john   32     1.8      65
print('df.iloc[[1]] =\n', df.iloc[[1]], sep='')

# df.iloc[1:,1:] =
#        height  weight
# john      1.8      65
# james     1.6      50
print('df.iloc[1:,1:] =\n', df.iloc[1:,1:], sep='')
```

### Accessed by name

```python
# retrieving as a pandas series
# df.loc['john'] =
# age       32.0
# height     1.8
# weight    65.0
print("df.loc['john'] =\n", df.loc['john'], sep='')

# retrieving as a pandas dataframe
# df.loc[['john']] =
#       age  height  weight
# john   32     1.8      65
print("df.loc[['john']] =\n", df.loc[['john']], sep='')

# df.loc['john':'james'] =
#        age  height  weight
# john    32     1.8      65
# james   45     1.6      50
print("df.loc['john':'james'] =\n", 
     df.loc['john':'james'], sep='')
```

# Data Cleaning

### Remove empty data

```python
df = pd.read_csv('data/data.csv')


print(df.shape[0])

# This will remove the null rows from the original data
# df.dropna(inplace = True)

# Identify rows with NaN values
rows_with_nan = df[df.isna().any(axis=1)]
print(rows_with_nan)

#Remove those with null values
new_df = df.dropna()
print(new_df.to_string())
#Print the total number of rows
print(new_df.shape[0])
```

### Replacing null values

```python
# Replace with a specified value
df_130 = rows_with_nan.fillna(130)
print(df_130.to_string())

#Find the mean of data.csv
mean_calories = df['Calories'].mean()
print('Mean calories: ',mean_calories)
#Replace with mean
df_mean = rows_with_nan.fillna(mean_calories)
print(df_mean.to_string())

#Find the median of data.csv
median_calories = df['Calories'].median()
print('Median calories: ',median_calories)
#Replace with median
df_median = rows_with_nan.fillna(median_calories)
print(df_median.to_string())

#Find the mode of data.csv
mode_calories = df['Calories'].mode()[0]
print('Mode calories: ',mode_calories)
#Replace with mode
df_mode = rows_with_nan.fillna(mode_calories)
print(df_mode.to_string())

```
